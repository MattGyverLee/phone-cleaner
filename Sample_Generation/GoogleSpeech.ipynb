```python
def text_to_ssml(input_text, split=False):
    ssml = []
    voice = True
    if split:
        for set in input_text:
            print(set)
            output = '<speak>'
            if voice:
                output += f"<voice name='{set[2]}'>"
            # Escape special characters in the text
            escaped_text = set[0].replace("&", "&amp;").replace("<", "&lt;").replace(">", "&gt;")
            escaped_simp = set[1].replace("&", "&amp;").replace("<", "&lt;").replace(">", "&gt;")
            output += f"<phoneme alphabet='ipa' ph='{escaped_text}'>{escaped_simp}</phoneme>"
            if voice:
                output += "</voice>"
            output += "</speak>"
            print(output)
            ssml.append(output)
    else:
        output = '<speak>'
        for set in input_text:
            print(set)
            # Escape special characters in the text
            escaped_text = set[0].replace("&", "&amp;").replace("<", "&lt;").replace(">", "&gt;")
            escaped_simp = set[1].replace("&", "&amp;").replace("<", "&lt;").replace(">", "&gt;")

            # Wrap the text in <speak> tags to create SSML
            if voice:
                output += f"<voice name='{set[2]}'>"
            output += f"<phoneme alphabet='ipa' ph='{escaped_text}'>{escaped_simp}</phoneme><break time='100ms'/>"
            if voice:
                output += "</voice>"

        output += "</speak>"
        print(output)
        ssml.append(output)    
    return ssml
```

```python
# Specify the input IPA string
ipa_strings = [["data","data","hi-IN-Neural2-A"], ["ɖata","ɖata", "hi-IN-Neural2-A"], ["ɗata","ɗata","hi-IN-Neural2-A"], ["d̪ata","d̪ata","hi-IN-Neural2-A"] ] 
language = "hi-IN"
split = True
ssml_input = text_to_ssml(ipa_strings, split)
# Create an input synthesis request


# Create a language code that is somewhat language-agnostic
language_code = language

# Specify the voice name for pronunciation (choose an appropriate voice)
voice_name = "hi-IN-Neural2-A"

# Create the voice selection
voice = texttospeech.VoiceSelectionParams(
    language_code=language_code, name=voice_name
)

# Configure the audio output parameters
audio_config = texttospeech.AudioConfig(
    audio_encoding=texttospeech.AudioEncoding.LINEAR16
)

print(ssml_input)
```

```python
# Perform the text-to-speech synthesis
for i, input in enumerate(ssml_input):
    print(input)
    file = ""
    if split:
        file = ipa_strings[i][0]   
    else:  
        first_strings = [sublist[0] for sublist in ipa_strings]
        file = "-".join(first_strings)

    synthesis_input = texttospeech.SynthesisInput(ssml=input)
    
    response = client.synthesize_speech(
        input=synthesis_input, voice=voice, audio_config=audio_config
    )
    
        # Save the synthesized audio to a file
    filename = "audio\\" + file + "-" + voice_name + ".wav"
    with open(filename, "wb") as out:
        out.write(response.audio_content)

    msg = f'Audio content written to "{filename}"'
```
